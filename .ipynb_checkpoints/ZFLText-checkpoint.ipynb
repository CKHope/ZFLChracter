```python
import numpy as np
import pandas as pd
```

```python
from chineseTextLib import *

fileName="zfl"
baseInputLink = f"./zfl/wordCountInput/{fileName}.txt"
dfWords=createWordListFromText(baseInputLink)
knownWords=createCharacterListFromText(baseInputLink)
# dfWords
# knownWords
```

```python
hanziDict=convertTxt2Dict(txtObject="dictionary.txt")
print(len(hanziDict))
```

```python
import json
hanziDictList=[]
for value in hanziDict.values():
    jsonValue=json.loads(value)
    hanziDictList.append(jsonValue)

# print(type(hanziDictList))
indexRange=np.arange(0,len(hanziDict))
# print(indexRange)
dfHanziDict=pd.DataFrame(hanziDictList,index=indexRange)
dfHanziDict.head(3)
```

```python
etymDict=dfHanziDict['etymology'].to_dict()
dfEtymDict=pd.DataFrame(etymDict).T
# print(len(dfEtymDict))
# print(dfEtymDict[:3])
dfHanziDict2=dfHanziDict.drop('etymology', axis=1)
dfHanziDict2
# print(len(dfHanziDict2))
# print(dfHanziDict2[:3])
dfNewTest=pd.concat([dfHanziDict2,dfEtymDict],axis=1)
dfNewTest.head(3)
```

```python
def saveDict2JSON(dict,outFileName="dict.json"):
    with open(outFileName, 'w') as outfile:
        json.dump(dict, outfile,indent=4)

saveDict2JSON(hanziDictList,"hanziDictList.json")
```

```python
df=dfNewTest
dfKradical=knownWords.merge(df,left_on="word",right_on="character")
dfKradical=dfKradical[['character','radical','times','definition','pinyin','type','phonetic','semantic','hint','decomposition']]    
dfKradical['percentile']=np.flip(np.arange(1,len(dfKradical)+1)/len(dfKradical)*100)

dictKradical=dfKradical.to_dict()
print(len(dictKradical['character']))
dfKradical.head(3)
```

```python
dictToProcess=dictKradical
print(max(dfKradical['times']))
tempDict=dfKradical.to_dict('records')
print(tempDict[:1])
from pyvis.network import Network
def map_data(characterData,shape='circle',charColor="white",radColor="#d08e21"):
    g=Network(notebook=True,height='900px',width="100%",bgcolor="black",font_color="black",
              select_menu=False,neighborhood_highlight=True,filter_menu=False)
    for node in characterData:
        if node['times']==1:
            node['color']="#ffffff"
            node['group']=4
        else:
            result=getColorBasedOnPercentile(node['percentile'])
            node['color']=result[0]
            node['group']=result[1]
    order=1
    for list in listOfRadList:
        # print(list)
        for node in characterData:
            if node['radical'] in list:
                #break all attributes of nodes to lines for easy control
                nodeTitle1=f"{node['pinyin']} | {node['decomposition']}\n {node['type']} | P: {node['phonetic']} -- S: {node['semantic']} \n"
                nodeTitle2=f"times--{node['times']} | hint--{node['hint']}\n meaning--{node['definition']}"
                g.add_node(node['character'],shape=shape,color=node['color'],title=nodeTitle1+nodeTitle2)
                g.add_node(node['radical'],shape=shape,color=radColor)
                g.add_edge(node['radical'],node['character'],color=node['color'])
        # g.repulsion(node_distance=80,central_gravity=0.2,spring_length=50,spring_strength=0.05,damping=0.09)
        # g.set_edge_smooth('diagonalCross')
        # g.show_buttons(True)
        g.set_options(options)
        g.show(f"N{order}--{list}.html",local=True)
        #reset network
        g=Network(notebook=True,height='900px',width="100%",bgcolor="black",font_color='black',
              select_menu=False,neighborhood_highlight=True,filter_menu=False)
        order+=1

# map_data(dictToProcess)
map_data(tempDict)
```
