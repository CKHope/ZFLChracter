## Asyncio request moedict -- multi request

```python
import asyncio
import httpx
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
import os

df=pd.read_pickle("dfCharacterZFL.pkl")
listWords=df['words'].tolist()

listOfWords=[]
for listItem in listWords:
    for i in range(len(listItem)):
        listOfWords.append(listItem[i])
listOfWords = sorted(set(listOfWords))
    
contentType = ("def", "example", "quote", "synonyms", "antonyms")

def updateResult(paragraphs, detailSpans: str, resultDict: dict) -> None:
    #loop through all spans in definition to get span type & content in definition
    allSpans = paragraphs.find_all("span", detailSpans)
    if allSpans != []:
        for s in allSpans:
            content = s.text
            # print(content)
            resultDict[detailSpans] = content        

async def getContent(client, url):
        resp = await client.get(url)
        content = resp.content
        
        return content

async def getChineseDetailsBatch():
    finalResult=[]
    async with httpx.AsyncClient() as client:
        #create a list of task for batch requests
        tasks = []
        for i in range(len(listOfWords)):
            url=f"https://www.moedict.tw/{listOfWords[i]}"
            tasks.append(asyncio.ensure_future(getContent(client, url)))
        gatherOfContent = await asyncio.gather(*tasks)
        
        #loop through task list result and get data
        for content in gatherOfContent:
            allResultDict={}
            singleResultDict = {}
            soup = bs(content, features="html.parser")
            definitionTag = soup.find_all("p", "definition")
            contentOrder = 1
            for definition in definitionTag:
                for i in range(len(contentType)):
                    updateResult(definition, contentType[i], singleResultDict)
                allResultDict[contentOrder]=singleResultDict
                singleResultDict={}
                contentOrder+=1
            finalResult.append(allResultDict)      
        print(len(tasks))
        return finalResult

print(len(listOfWords))
# listOfWords

fileName="wordChineseDetailsDataZFL.pkl"
if os.path.exists(fileName):
    print(f"File {fileName} already created!")
    dfwordDetails=pd.read_pickle(fileName)
    dfwordDetails.columns=['word','wordChineseDetails']
    pass
else:
    # await getChineseDetailsBatch()
    listOfChineseDetail=await getChineseDetailsBatch()
    # listOfChineseDetail[:5]
    resultList=[list(l) for l in zip(listOfWords,listOfChineseDetail)]
    dfwordDetails=pd.DataFrame(resultList)
    dfwordDetails.columns=['word','wordChineseDetails']
    dfwordDetails.to_pickle(fileName)
    
dfwordDetails[:3]
```
